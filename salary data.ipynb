```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('salary.csv')
data
```

```python
ds=pd.DataFrame(data=data)
```

```python
ds
```

```python
sns.heatmap(ds.isnull())
```

```python
ds.isnull().sum()
```

```python
ds['City']
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='most_frequent')
ds['City']=imp.fit_transform(ds['City'].values.reshape(-1,1))
ds
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='mean')
ds['Age']=imp.fit_transform(ds['Age'].values.reshape(-1,1))
ds['Salary']=imp.fit_transform(ds['Salary'].values.reshape(-1,1))
ds
```

```python
ds.drop(['Name'],axis=1,inplace=True)
```

```python
ds
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['City']=le.fit_transform(ds['City'])
ds['Country']=le.fit_transform(ds['Country'])
ds
```

```python
x=ds.iloc[:,0:4]
x
```

```python
y=ds.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=70)
lm=LinearRegression()
lm.fit(x_train,y_train)
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
pred
```

```python
print(r2_score(y_test,pred))
```
